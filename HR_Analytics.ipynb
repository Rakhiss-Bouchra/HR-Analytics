### Importing Libraries

```python
# Count the number of unique values in each column
Data.nunique()
```

```python
# displaying columns with constant values
constant_columns = ["EmployeeCount", "Over18", "StandardHours"]

for column in constant_columns:
    unique_values = Data[column].unique()
    print(f"{column}: {unique_values}")
```

It seems that the columns "EmployeeCount", "Over18", and "StandardHours" have only one unique value each. In most cases, these columns do not provide any meaningful information for analysis because they are constant, so we can consider dropping them a cleaning of the dataset and to improve the efficiency of the analysis!

Dropping duplicates if they exist!

```python
# Display the number of duplicate rows
num_duplicates = Data.duplicated().sum()
print(f"Number of Duplicate Rows: {num_duplicates}")
```

##### The correlation matrix

Before deleting, we must know relationship between columns!

```python
# Select only numeric columns for correlation
numeric_columns = Data.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix for numeric columns
correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5)
plt.title("Correlation Map for Numeric Variables")
plt.show()
```

```python
# List of columns to drop
columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']

# Drop the specified columns
Final_Data = Data.drop(columns=columns_to_drop)
```

Renaming the columns

```python
# Renaming selected columns to improve clarity
Final_Data = Final_Data.rename(columns={'BusinessTravel': 'Business Travel' , 'DailyRate': 'Daily Rate', 'DistanceFromHome' : 'Distance From Home', 'EducationField' : 'Education Field', 
                                'EmployeeNumber' : 'Employee Number', 'EnvironmentSatisfaction' : 'Environment Satisfaction' , 'HourlyRate' : 'Hourly Rate', 
                                'JobInvolvement' : 'Job Involvement' , 'JobLevel' : 'Job Level' , 'JobRole' : 'Job Role', 'JobSatisfaction' : 'Job Satisfaction', 
                                'MaritalStatus' : 'Marital Status', 'MonthlyIncome' : 'Monthly Income', 'MonthlyRate' : 'Monthly Rate', 'NumCompaniesWorked' : 'Num Companies Worked', 
                                'OverTime' : 'Over Time', 'PercentSalaryHike' : 'Percent Salary Hike' , 'PerformanceRating' : 'Performance Rating', 
                                'RelationshipSatisfaction' : 'Relationship Satisfaction', 'StockOptionLevel' : 'Stock Option Level', 'TotalWorkingYears' : 'Working Years', 
                                'TrainingTimesLastYear' : 'Training Times Last Year', 'WorkLifeBalance' : 'Work Life Balance', 'YearsAtCompany' : 'Years At Company' , 
                                'YearsInCurrentRole' : 'Years In Current Role', 'YearsSinceLastPromotion' : 'Years Since Last Promotion', 'YearsWithCurrManager' : 'Years With Current Manager'})
```

### Data Visualization

```python
# Visualize the gender distribution
gender_count = Final_Data['Gender'].value_counts()
keys = ['Male', 'Female']
palette_color = sns.color_palette("Paired")
plt.pie(gender_count, labels=keys, colors=palette_color, autopct='%.0f%%', center=(0, 0), explode=[0.1, 0] )
```

The dataset is predominantly composed of male employees, accounting for approximately 60% of the total. Female employees, on the other hand, make up the remaining 40%. 

```python
# Visualize the relationship between Department, Hourly Rate, and Over Time
plt.figure(figsize=(10,5))
sns.barplot(data=Final_Data ,hue="Over Time", x="Department", y="Hourly Rate", palette="Paired", width=0.4, ci=None)
```

As we can observe, the hourly rates for each department are approximately similar, with a slight variation. Notably, the "Research & Development" department exhibits a slightly higher hourly rate than the others.  
When considering the **Over Time** variable, the "Sales" department has the highest occurrence, followed by the "Research & Development" department, and lastly, the "Human Resources" department.

```python
# Visualize the relationship between Marital Status and Over Time
plt.figure(figsize=(10, 5))
sns.countplot(data=Final_Data, x="Marital Status", hue="Over Time", palette="Paired", width=0.4)
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()
```

From the chart, it is evident that the majority of employees are married. Interestingly, regardless of their marital status, a significant number of employees do not work overtime.  
This suggests that overtime work is not highly prevalent among the employees, irrespective of their marital status.

```python
# Visualize the relationship between Department and Job Satisfaction
plt.figure(figsize=(15, 5))

# Countplot for Department
plt.subplot(1, 2, 1)
sns.countplot(data=Final_Data, x="Department", palette="Paired", width=0.4)
plt.title("Department Count")
# Countplot for Department with Job Satisfaction
plt.subplot(1, 2, 2)
sns.countplot(data=Final_Data, x="Department", hue="Job Satisfaction", palette="Paired", width=0.4)
plt.title("Department Count with Job Satisfaction")
plt.tight_layout()
plt.show()
```

First, when analyzing the department-wise distribution of employees, it becomes evident that the 'Research and Development' department has a significantly higher number of employees, followed by the 'Sales' department, and with the 'Human Resources' department having the fewest employees.  

When considering job satisfaction within these departments, the data reveals interesting insights. In the 'Research and Development' department, most employees appear to be moderately satisfied, with job satisfaction ratings predominantly falling within the range of 3 (slightly satisfied) and 4 (satisfied). Similarly, in the 'Sales' department, employees also show moderate satisfaction, with a slightly higher preference for a rating of 4.  
However, the situation differs in the 'Human Resources' department, where the job satisfaction ratings tend to be lower, with a higher prevalence of ratings at 2 (slightly dissatisfied).

```python
# Distribution of employees with high and lowest job satisfaction
filtered_data_1 = Final_Data[Final_Data['Job Satisfaction'] == 1]
percentage_data_1 = filtered_data_1.groupby('Job Role').size() / Final_Data.groupby('Job Role').size() * 100

filtered_data_4 = Final_Data[Final_Data['Job Satisfaction'] == 4]
percentage_data_4 = filtered_data_4.groupby('Job Role').size() / Final_Data.groupby('Job Role').size() * 100

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].pie(percentage_data_4, labels=percentage_data_4.index, autopct='%1.1f%%', startangle=140)
axes[0].axis('equal')
axes[0].set_title('Distribution of employees with high job satisfaction "4"')

axes[1].pie(percentage_data_1, labels=percentage_data_1.index, autopct='%1.1f%%', startangle=140)
axes[1].axis('equal')  
axes[1].set_title('Distribution of employees with lowest job satisfaction "1"')
plt.tight_layout()
plt.show()
```

The chart shows the distribution of employees with high and lowest job satisfaction (rated 4 & 1) across different Job Roles:

- The Job Role with the highest percentage of employees who have a Job Satisfaction rating of **4** is **Sales Executive**, accounting for **12.8%** of such employees.
- The Job Role with the highest percentage of employees who have a Job Satisfaction rating of **1** is **Laboratory Technician**, representing **12.6%** of such employees.

```python
# The employee distribution across different job roles
plt.figure(figsize=(10, 5))
sns.relplot(data=Final_Data, kind="line",x="Job Role", y="Employee Number",  height=5, aspect=2)
plt.xticks(rotation=45)
```

The **Human Resources** department has the highest number of employees, while the department with the lowest number of employees is **Manager**.

```python
# How education levels and monthly incomes are distributed across different job roles
sns.set_style("darkgrid")
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# First subplot: Barplot for Education
sns.barplot(data=Final_Data[Final_Data['Education'] > 1], x='Job Role', y='Education', ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)
axes[0].set_title('Education Level by Job Role')

# Second subplot: Barplot for Monthly Income
sns.barplot(data=Final_Data[Final_Data['Monthly Income'] > 3000], x='Job Role', y='Monthly Income', ax=axes[1])
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
axes[1].set_title('Monthly Income by Job Role')
plt.tight_layout()
plt.show()
```

In the first subplot, the highest bar corresponds to the **Research Director** job role, while the lowest bar represents the **Sales Representative** job role.  
In the second subplot, the tallest bar represents the **Manager** job role.

```python
sns.set_style("darkgrid")
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.barplot(data=Final_Data, x = 'Department', y = 'Education', ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)

sns.barplot(data=Final_Data[Final_Data['Education']>2] ,x = 'Education', y = 'Monthly Income', ax=axes[1])
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
plt.tight_layout()
plt.show()
```

```python
data_subset = Final_Data[['Business Travel', 'Marital Status']]
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
marital_statuses = data_subset['Marital Status'].unique()

for i, status in enumerate(marital_statuses):
    subset_data = data_subset[data_subset['Marital Status'] == status]
    counts = subset_data['Business Travel'].value_counts()
    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%')
    axes[i].set_title(f'{status}')

plt.suptitle('Relation entre Business Travel et Marital Status')
plt.show()
```
